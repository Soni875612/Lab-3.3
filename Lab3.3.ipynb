```python
# import libraries 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew, kurtosis
```

```python
# load the dataset
df = pd.read_csv("GOI.csv")
df.head()
```

```python
# Extract Rate data from the Literacy
literacy_rates = df["Literacy Rate (Persons) - Total - 2011"].dropna()
```

```python
# Compute Skewness and Kurtosis
skewness = skew(literacy_rates)
kurt = kurtosis(literacy_rates)
```

```python
# Print Results
print(f"Skewness: {skewness}")
print(f"Kurtosis: {kurt}")
```

```python
# Visualization using Histogram
plt.figure(figsize=(8,5))
sns.histplot(literacy_rates, kde=True, bins=15, color="blue")
plt.title("Distribution of Literacy Rates")
plt.xlabel("Literacy Rate (%)")
plt.ylabel("Frequency")
plt.show()
```
